### 영화

데이터: TMDB 5000 movie dataset from Kaggle

```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

movies = pd.read_csv('https://raw.githubusercontent.com/PinkWink/ML_tutorial/master/dataset/tmdb_5000_movies.csv')
print(movies.shape)
```

```python
movies.head()
```

```python
# 데이터 선택

movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']]
movies_df.head()
```

```python
# genres, keywords 컬럼은 list내의 dict 모양의 str로 저장되어있음

movies_df['genres'][0], type(movies_df['genres'][0])
```

```python
# literal_eval을 이용하면 파이썬에서 사용하는 대로의 형태를 적용하게 됨
# genres, keywords str >> list/dict로 변경


from ast import literal_eval

movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
movies_df.head()
```

```python
# 변경된 데이터 타입 확인

movies_df.genres[0]
```

```python
[y['name'] for y in movies_df['genres'][0]]
```

```python
# 위에서 dictionary 값 적용한 것을 각 컬럼에 적용

movies_df.genres = movies_df.genres.apply(lambda x : [y['name'] for y in x])
movies_df.keywords = movies_df.keywords.apply(lambda x : [y['name'] for y in x])
movies_df[['genres', 'keywords']][:5]
```

```python
# test

(' ').join(movies_df['genres'][0])
```

```python
# genres를 띄어쓰기로 구분한 형식의 컬럼 생성

movies_df['genres_literal'] = movies_df.genres.apply(lambda x: (' ').join(x))
movies_df.head(3)
```

```python
# countvectorize 수행

from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df = 0.0, ngram_range=(1, 2)) # ngram_range >> 단어 조합 최소, 최대 값
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

* [ngram_range 이해](https://saturncloud.io/blog/understanding-the-ngramrange-argument-in-a-countvectorizer-in-sklearn/)

```python
# 문장의 유사도 측정하는 방법 중 하나 인 코사인 유사도 측정 수행
# 반비례 관계 = -1, 관계 없음 = 0, 비례 관계 = 1

from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:2])
```

```python
# genre_sim index 순서대로 정렬

genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1] # ::-1 >> numpy에서 정렬하는 방식
print(genre_sim_sorted_ind[:1])
```

```python
# 추천 영화를 dataframe으로 반환하는 함수 작성

def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :(top_n)]

    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]
```

```python
# The Godfather 와 유사한 영화 찾아보기

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average']]
```

* Mi America 같은 경우 평점이 0.0임
* 데이터 재확인이 필요

```python
# vote count와 함께 데이터 보기

movies_df[['title', 'vote_average', 'vote_count']].sort_values('vote_average', ascending=False)[:10]
```

* 1~2명이 평점을 매긴 영화는 해당 평점에 신뢰도가 낮음
* 고로 가중치 선정
$$(\frac{v}{v+m})R + (\frac{m}{v+m})C$$

* v: 개별 영화에 평점을 투표한 횟수
* m: 평점을 부여하기 위한 최소 투표 횟수
* R: 개별 영화에 대한 평균 평점
* C: 전체 영화에 대한 평균 평점

```python
# 위 가중치 적용

C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) # 최소 투표 횟수는 임의 선정

C, m
```

```python
# 가중치 부여된 평점을 계산하기 위한 함수 작성

def weighted_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    return ((v/(v+m))*R) + ((m/(m+v))*C)
```

```python
# 가중치 적용된 평점 컬럼 생성

movies_df['weighted_vote'] = movies_df.apply(weighted_vote_average, axis=1)
movies_df.head()
```

```python
# 추천 영화를 dataframe으로 반환하는 함수 수정

def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, :top_n*2] # 애초에 10개만 뽑은 것 보다 더 많은 영화를 가져와야 밑에서 가중치에 따라 정렬이 가능함

    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    similar_indexes = similar_indexes[similar_indexes != title_index] # 추천 영화를 위해 input으로 넣은 영화 제거

    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:10]
```

```python
# 다시 한 번 The Godfather 와 유사한 영화 찾아보기

similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```

### MovieLens Data (아이템 기반 최근접 이웃 협업 필터링)

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('../ML_data/ml-latest-small/movies.csv')
ratings = pd.read_csv('../ML_data/ml-latest-small/ratings.csv')

print(movies.shape)
print(ratings.shape)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
# pivot_table을 통해 두 테이블 조합

ratings = ratings[['userId', 'movieId', 'rating']]
ratings_matrix = ratings.pivot_table('rating', index = 'userId', columns='movieId')
ratings_matrix.head()
```

* movieId 대신에 movie title을 넣기 위해 테이블 두 개를 다시 합쳐서 pivot table 재이용

```python
# ratings + movies (movie Id 기준으로 결합)

rating_movies = pd.merge(ratings, movies, on='movieId')
rating_movies.head()
```

```python
# 다시 pivot table 이용하여 변환

ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')
ratings_matrix.head()
```

```python
# NaN 값은 0으로 변환

ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head()
```

```python
# 유사도 측정을 위해 행렬 transpose

ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head()
```

```python
# 유사도 측정 및 결과 확인

from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df = pd.DataFrame(data=item_sim, index = ratings_matrix.columns,
                           columns = ratings_matrix.columns)
print(item_sim_df.shape)
item_sim_df.head()
```

```python
# 대부와 유사한 영화 확인

item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[1:11]
```

```python
# 데이터 경로 확인

import numpy as np
import pandas as pd

import os
print(os.listdir('../ML_data/goodbooks_10k/'))
```

```python
# ratings.csv 가져오기

ratings = pd.read_csv('../ML_data/goodbooks_10k/ratings.csv')
ratings.head()
```

```python
# tags.csv 가져오기

tags = pd.read_csv('../ML_data/goodbooks_10k/tags.csv')
tags.tail()
```

```python
# book_tag, tag merge

tags_join_df = pd.merge(book_tags, tags, left_on='tag_id', right_on='tag_id', how='inner')
tags_join_df.head()
```

* left_on : 좌측에 붙이는 dataframe을 어떤 컬럼에 키를 맞출 것인지 정하는 것
* right_on : 우측에 붙이는 dataframe을 어떤 컬럼에 키를 맞출 것인지 정하는 것

```python
# books 의 저자 확인

books.authors[:5]
```

```python
# authors로 Tfidf 수행

from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0.0, stop_words='english')

tfidf_matrix = tf.fit_transform(books['authors'])
tfidf_matrix
```

```python
# 유사도 측정

from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
# The Hobbit의 index 찾기

titles = books['title']
indices = pd.Series(books.index, index = titles)
indices['The Hobbit']
```

```python
# The Hobbit의 유사 값 추출

cosine_sim[indices['The Hobbit']]
```

```python
# 유사도 결과를 인덱스를 가진 list 형으로 변경

cosine_sim[indices['The Hobbit']].shape
```

```python
# 1만개의 책과 비교하였을 때, 마지막 5개의 책들과 유사성 비교 결과

list(enumerate(cosine_sim[indices['The Hobbit']]))[-5:]
```

```python
# 가장 유사한 책 10권의 인덱스 찾기

sim_scores = list(enumerate(cosine_sim[indices['The Hobbit']]))
sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True) # 두번째 원소에 대해 descending 하도록 sorting
sim_scores[1:11]
```

```python
# 작가로 본 유사 책 검색

sim_scores = sim_scores[1:11]
book_indices = [i[0] for i in sim_scores] # tuple 형태, 각 원소 첫번째가 index임
titles.iloc[book_indices] # 위에서 만들었던 books['title'] 변수 활용
```

```python
# book에 tag 포함

books_with_tags = pd.merge(books, tags_join_df, left_on = 'book_id', right_on='goodreads_book_id', how='inner')
books_with_tags.head()
```

```python
# tag로 Tfidf 수행

tf1 = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0.0, stop_words='english')
tfidf_matrix1 = tf1.fit_transform(books_with_tags['tag_name'].head(10000))
cosine_sim1 = linear_kernel(tfidf_matrix1, tfidf_matrix1)
```

```python
# 위에서 활용한 변수들을 조합하여 추천책 반환하는 함수 작성

titles1 = books['title']
indices1 = pd.Series(books.index, index=titles1)

def tags_recommendation(title):
    idx = indices1[title]
    sim_scores = list(enumerate(cosine_sim1[idx]))
    sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    book_indices = [i[0] for i in sim_scores]
    return titles1.iloc[book_indices]
```

```python
tags_recommendation('The Hobbit')
```

```python
# tag name이 모두 따로따로 떨어져있으므로 bookid에 대해 groupby로 결합

temp_df = books_with_tags.groupby('book_id')['tag_name'].apply(' '.join).reset_index()
temp_df.head()
```

```python
# 전체 결합한 tag들을 books dataframe에 추가

books = pd.merge(books, temp_df, left_on = 'book_id', right_on = 'book_id', how='inner')
books.head()
```

```python
# 저자 이름도 tag에 합치기

books['corpus'] = (pd.Series(books[['authors', 'tag_name']].fillna('').values.tolist()).str.join(' '))
# 두 컬럼의 데이터를 가져오되, NaN 값은 빈칸으로 바꾸고, 값들을 리스트로 변환 >> 이후 str 변환하여 합친다는 뜻

books.corpus[:3]
```

```python
# corpus에 대해 tfidf 수행

tf_corpus = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df = 0.0, stop_words='english')
tfidf_matrix_corpus = tf_corpus.fit_transform(books['corpus'])
cosine_sim_corpus = linear_kernel(tfidf_matrix_corpus, tfidf_matrix_corpus)

titles = books['title']
indices = pd.Series(books.index, index = titles)
```

```python
# tags 추천 함수 수정 보완

def corpus_recommendation(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim_corpus[idx]))
    sim_scores = sorted(sim_scores, key=lambda x : x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    book_indices = [i[0] for i in sim_scores]
    return titles1.iloc[book_indices]
```

```python
corpus_recommendation('The Hobbit')
```
