```python
import pycountry
```

```python
# pycountryì™€ forë¬¸ ì´ìš©í•˜ì—¬ ê²€ìƒ‰ ì•ˆë˜ëŠ” êµ­ê°€ í™•ì¸

country_id = []

for idx, rows in df_target.iterrows():
    country_name = df_target.country[idx]
    country = pycountry.countries.get(name=country_name)
    country_id.append([idx, country])

country_id
```

```python
# 2ê¸€ì ì½”ë“œ ë¹¼ëŠ” ë²•

country_id[0][1].alpha_2
```

```python
# pycountry ì¼ë°˜ ê²€ìƒ‰ ì•ˆë˜ëŠ” êµ­ê°€ ë”°ë¡œ ì¶”ì¶œ

country_none = []

for i in country_id:
    if i[1] == None:
        country_none.append(i)

country_none
```

```python
# fuzzy search
country_id2 = []
for i in country_none:
    country_name = df_target.country[i[0]]
    country = pycountry.countries.search_fuzzy(country_name)
    country_id2.append([i[0], country])

country_id2

# ëª¨ë‘ ë‹¤ì¤‘ ë¦¬ìŠ¤íŠ¸ í˜•ì‹ìœ¼ë¡œ ë˜ì–´ìˆì–´ì„œ ì²«ë²ˆì§¸ë§Œ ê°€ì ¸ì˜¤ë©´ ë˜ëŠ” ìƒí™©ì—ì„œ êµ³ì´ iran ê²ƒì„ ë³€ê²½í•  í•„ìš”ëŠ” ì—†ìŒ
```

```python
# code ì—´ ìƒì„±
import numpy as np

df_target['code'] = np.nan
df_target.head()
```

```python
# ë§Œë“¤ì–´ì§„ ë‘ ë¦¬ìŠ¤íŠ¸ë¥¼ ì´ìš©í•˜ì—¬ df_target ë°ì´í„° í”„ë ˆì„ì— ê¸°ì…

# pycountry.countries.get ì´ìš©í•œ ì •ë³´ ê¸°ì…
for i in country_id:
    try:
        df_target.code[i[0]] = i[1].alpha_2
    except:
        continue

# pycountry.countries.fuzzy_search ì´ìš©í•œ ì •ë³´ ê¸°ì…
for x in country_id2:
    df_target.code[x[0]] = x[1][0].alpha_2

df_target.head()
```
