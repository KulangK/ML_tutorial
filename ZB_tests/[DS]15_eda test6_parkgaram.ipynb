````python
import sympy as sy
import numpy as np
````

````python
# sympy test

x = sy.symbols('x')

f = sy.Eq(df_target.여자[0] * x/100 + df_target.남자[0] * (1 - x/100), df_target.전체[0])

result = sy.solve(f)
result
````

````python
# sympy 적용하여 각 행에 결과 대입

x = sy.symbols('x')
df_target['여자비율'] = np.nan

for idx, rows in df_target.iterrows():
    try:
        f = sy.Eq(df_target.여자[idx] * x/100 + df_target.남자[idx] * (1 - x/100), df_target.전체[idx])
        df_target['여자비율'][idx] = sy.solve(f)[0]
    except:
        df_target['여자비율'][idx] = None
````

````python
# sympy 결과는 float64가 아니기 때문에 데이터 타입 변환 및, 반올림

df_target.여자비율 = df_target.여자비율.astype('float64').round(6)
df_target.head()
````
