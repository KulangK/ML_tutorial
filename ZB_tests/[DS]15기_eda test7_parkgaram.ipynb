## 순위 증감 분리

```python
# for문 이용하여 순위와 순위 증감 분리하기

rank = []
rank_comp = []

for i in range(len(ranking_raw)):
    a = ranking_raw['Unnamed: 0'][i]
    b = re.sub('\(\s|\s\)|\(|\)', '', a).split('  ')
    rank.append(b[0])
    try:
        rank_comp.append(b[1])
    except:
        rank_comp.append('new')

rank, rank_comp
```

```python
# 20210812(목)~20220724(일)
# child 57~106

for i in range(57, 107, -1):
    # 페이지 변경
    find_date = driver.find_element(By.CSS_SELECTOR, f'body > table > tbody > tr:nth-child(2) > td:nth-child(3) > table > tbody > tr > td > table:nth-child(4) > tbody > tr:nth-child(2) > td > table:nth-child(1) > tbody > tr:nth-child(3) > td > form > select > option:nth-child({i})')
    find_date.click() # 2022.07.21(목)~2022.07.24(일) 일자 찾아서 클릭

    time.sleep(1.0)

    # 페이지 정보 획득
    html = driver.page_source

    # 주 정보 가져오기
    dom = BeautifulSoup(html, 'html.parser')

    detail = dom.select(f'body > table > tbody > tr:nth-child(2) > td:nth-child(3) > table > tbody > tr > td > table:nth-child(4) > tbody > tr:nth-child(2) > td > table:nth-child(1) > tbody > tr:nth-child(3) > td > form > select > option:nth-child({i})')
    a = detail[0].get('value')
    
    # 데이터프레임 만들기
    table = pd.read_html(html, header=0)
    ranking_raw = table[-11]
    ranking_raw['주'] = a

    time.sleep(1.0)

    if i == 57:
        ranking_tot = ranking_raw
    else:
        ranking_tot = pd.concat([ranking_tot, ranking_raw])
    
    print(len(ranking_tot))
```

```python
# 데이터 프레임 확인 및 테스트용 카피본 만들기

test = ranking_tot.copy()
ranking_tot.head()
```

```python
test = test.drop(['Unnamed: 3', 'Unnamed: 5'], axis=1)
test = test.dropna().reset_index(drop=True)
test.head()
```

```python
test['Unnamed: 4'] = test['Unnamed: 4'].str.replace('.', '')
test['Unnamed: 4'] = test['Unnamed: 4'].astype('int')
test['주'] = test['주'].astype('int')
test.head()
```

```python
# for문 이용하여 순위와 순위 증감 분리하기

rank = []
rank_comp = []

for i in range(len(test)):
    a = test['Unnamed: 0'][i]
    b = re.sub('\(\s|\s\)|\(|\)', '', a).split('  ')
    rank.append(b[0])
    try:
        rank_comp.append(b[1])
    except:
        rank_comp.append('new')
```

```python
# 위에서 만든 순위와 순위증감 추가

test['순위'] = rank
test['순위증감'] = rank_comp
test.head()
```

```python
# 데이터프레임 확인

df_target.head()
```

```python
df_test = df_target.copy()
```

```python
# % 기호 없애기

df_test.예매점유율 = df_test.예매점유율.str.replace('%', '')
df_test.head()
```

```python
# 예매점유율 str > float 변환

df_test.예매점유율 = df_test.예매점유율.astype('float')
df_test.info()
```

```python
# 예매 점유율 평균 계산

df_test = df_test.groupby('영화제목').agg({'예매점유율':'mean'}).round(2)

df_test
```

```python
# 예매점유율 내림차순, 영화제목 오름차순 정렬 후 소수점 2째자리까지 표현

df_test = df_test.sort_values(by=['예매점유율', '영화제목'], ascending=[False, True])
df_test['예매점유율'] = df_test['예매점유율'].apply(lambda x: '{0:.2f}'.format(x))
df_test
```

```python
# % 표시 붙이기

df_test.예매점유율 = df_test.예매점유율.astype(str) + '%'
df_test
```
