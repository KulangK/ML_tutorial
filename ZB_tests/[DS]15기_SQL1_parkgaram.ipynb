````python
# 스타벅스 페이지 접근
from selenium import webdriver

url = "https://www.starbucks.co.kr/store/store_map.do"
driver = webdriver.Chrome('../driver/chromedriver.exe')
driver.get(url)


# 팝업 창 닫기 예외 처리
try:
    driver.find_element_by_css_selector('.holiday_notice_close a').click()

except Exception as e:
    print(str(e)) # str() 없이 그냥 e쓰면 웹페이지에 예상외의 문구 뜸


# 지역 검색 클릭, 서울 선택, 전체 선택
driver.find_element_by_css_selector('#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a').click()

driver.find_element_by_css_selector('#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a').click()

time.sleep(0.5)

driver.find_element_by_css_selector('#mCSB_2_container > ul > li:nth-child(1) > a').click()

time.sleep(0.5)


# 전체 리스트 가져오기
seoul_list_sel = driver.find_elements_by_css_selector('#mCSB_3_container ul li')

html = driver.page_source
dom = BeautifulSoup(html, 'html.parser')

seoul_list = dom.select('#mCSB_3_container ul li')
````

````python
# 매장 이름
seoul_list[0]['data-name']

# 매장이 위치한 구 이름
seoul_list[0].select_one('p').text[:-9].split()[1]

# 매장 주소
seoul_list[0].select_one('p').text[:-9]

# 위도
seoul_list[0]['data-lat']

# 경도
seoul_list[0]['data-long']
````

````python
# 테이블에 데이터 넣기

remote = mysql.connector.connect(
    host = "oneday.cwbv9r9iyiht.us-east-2.rds.amazonaws.com",
    port = 3306,
    user = "oneday",
    password = "12341234",
    database = "oneday"
)

cur = remote.cursor()
sql = "insert into COFFEE_STORE (id, brand, name, gu_name, address, lat, lng)\
        values (%s, %s, %s, %s, %s, %s, %s)"
for x in range(len(driver.find_elements_by_css_selector('#mCSB_3_container ul li'))):
    data = (x+1, # id
            1, # brand numbering
            seoul_list[x]['data-name'], # 매장 이름
            seoul_list[x].select_one('p').text[:-9].split()[1], # 매장 행정 구역
            seoul_list[x].select_one('p').text[:-9], # 매장 주소
            seoul_list[x]['data-lat'], # 매장 위도
            seoul_list[x]['data-long']) # 매장 경도
    cur.execute(sql, tuple(data))
    print(data)
    remote.commit()  # 최초 commit 이 필요한 것을 깜빡하여 데이터가 계속 안들어갔었음.

cur.execute("select * from COFFEE_STORE")
result = cur.fetchall()

remote.close()
````

- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다. 

````python
# MySQL workbench 이용하여 다양한 조합 사용해보았음

remote = mysql.connector.connect(
    host = "oneday.cwbv9r9iyiht.us-east-2.rds.amazonaws.com",
    port = 3306,
    user = "oneday",
    password = "12341234",
    database = "oneday"
)

cur = remote.cursor()
cur.execute('select s.id s_id, b.name s_brand, s.name s_name, s.gu_name s_gu, s.address s_address, s.lat s_lat, s.lng s_lng, \
                    e.id e_id, d.name e_brand, e.name e_name, e.gu_name e_gu, e.address e_address, e.lat e_lat, e.lng e_lng \
            from (select * from COFFEE_STORE where brand = 1) s, \
                 (select * from COFFEE_STORE where brand = 2) e, \
	             (select * from COFFEE_BRAND where id = 1) b, \
                 (select * from COFFEE_BRAND where id = 2) d \
            where s.brand = b.id \
            order by s_id, e_id')

result = cur.fetchall()
df = pd.DataFrame(result, columns = cur.column_names)

remote.close()
````
