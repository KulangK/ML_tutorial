- 주소를 입력받아 구 이름을 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
- 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)

````python
# 화폐단위 문자형을 입력 받아 숫자형으로 반환하는 함수

def change(x):
    x = x.replace(',', '')
    x = int(x)
    return x

change('1,000')
````

````python
# 주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID를 반환하는 함수

def brand_Id(x):
    brand = (x,)
    remote = mysql.connector.connect(
        host = "oneday.cwbv9r9iyiht.us-east-2.rds.amazonaws.com",
        port = 3306,
        user = "oneday",
        password = "12341234",
        database = "oneday"
    )
    cur = remote.cursor()
    sql = 'select id from GAS_BRAND where name = %s'
    cur.execute(sql, brand)
    tmp = cur.fetchall()
    result = [i[0] for i in tmp]
    remote.close()
    return result[0]

brand_Id('SK에너지')
````

````python
# 주소를 입력 받아 행정 구를 반환하는 함수

def gu(x):
    y = x.split()[1]
    return y

gu('서울시 강남구 헌릉로 730')
````

````python
# 주소를 입력받아 위도, 경도를 반환하는 함수

def lat_lng(x):
    ll = []
    tmp = gmaps.geocode(x, language='ko')
    ll.append(tmp[0].get('geometry')['location']['lat'])
    ll.append(tmp[0].get('geometry')['location']['lng'])
    return ll

lat_lng('서울시 강남구 헌릉로 730')
````

````python
remote = mysql.connector.connect(
    host = "oneday.cwbv9r9iyiht.us-east-2.rds.amazonaws.com",
    port = 3306,
    user = "oneday",
    password = "12341234",
    database = "oneday"
)

cur = remote.cursor()
sql = 'insert into GAS_STATION (brand, name, city, gu, address, gasoline, diesel, self, car_wash, charging_station, car_maintenance, convenience_store, 24_hours, lat, lng) \
       values (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'

y = 0 # total number 확인하기 위함

for i1 in gu_names:
    driver.find_element_by_css_selector('#SIGUNGU_NM0').send_keys(i1)  # 구 선택
    x = int(driver.find_element_by_css_selector('#totCnt').text)            # 해당 구 주유소 갯수 (내부 for문에 사용)
    body1 = driver.find_element_by_css_selector('#body1')              # 추후 driver 사용 축약 위함.
    for i2 in range(1, int(x)+1):
        driver.find_element_by_css_selector(f'#body1 > tr:nth-child({i2}) > td.rlist > a').click() # 각 주유소 클릭
        br_tmp = body1.find_elements_by_css_selector('td.rlist img')[i2-1].get_attribute('alt')
        if br_tmp == '알뜰':
            brand = '알뜰주유소'
        elif br_tmp == 'PB':
            brand = '자가상표'
        else:
            brand = br_tmp
        name = driver.find_element_by_css_selector('.header #os_nm').get_attribute('innerText') # innerText > 화면에 안보이는 텍스트 가져옴 (생략된 것들)
        city = '서울'
        gu = i1
        address = driver.find_element_by_css_selector('#rd_addr').get_attribute('innerText')
        gas = body1.find_elements_by_css_selector('td:nth-child(2)')[i2-1].text
        diesel = body1.find_elements_by_css_selector('td:nth-child(3)')[i2-1].text
        try:
            driver.find_element_by_css_selector('#self_icon').get_attribute('alt')
            self = 1
        except:
            self = 0
        car_wash = 0 if 'off' in driver.find_element_by_css_selector('.service #cwsh_yn').get_attribute('src') else 1
        charge = 0 if 'off' in driver.find_element_by_css_selector('.service #lpg_yn').get_attribute('src') else 1
        maint = 0 if 'off' in driver.find_element_by_css_selector('.service #maint_yn').get_attribute('src') else 1
        conve = 0 if 'off' in driver.find_element_by_css_selector('.service #cvs_yn').get_attribute('src') else 1
        hours = 0 if 'off' in driver.find_element_by_css_selector('.service #sel24_yn').get_attribute('src') else 1
        lat = lat_lng(address)[0]
        lng = lat_lng(address)[1]
        data = (brand_Id(brand), name, city, gu, address, gas, diesel, self, car_wash, charge, maint, conve, hours, lat, lng)
        cur.execute(sql, tuple(data))
        remote.commit()
        print(data)
    y += x
    time.sleep(0.7)

cur.execute('select * from GAS_STATION')
result = cur.fetchall()

remote.close
````

````python
remote = mysql.connector.connect(
    host = "oneday.cwbv9r9iyiht.us-east-2.rds.amazonaws.com",
    port = 3306,
    user = "oneday",
    password = "12341234",
    database = "oneday"
)

cur = remote.cursor()

cur.execute('select s.id id, b.name brand, s.name name, s.city city, s.gu gu, s.address address, s.gasoline gasoline, \
	     s.diesel diesel, s.self self, s.car_wash car_wash, s.charging_station charging_station, s.car_maintenance car_maintenance,\
            s.convenience_store convenience_store, s.24_hours 24_hours, s.lat lat, s.lng lng\
            from GAS_BRAND b, GAS_STATION s\
            where b.id = s.brand\
            order by id')

result = cur.fetchall()
df = pd.DataFrame(result, columns = cur.column_names)

remote.close()
````

       s.self self, s.24_hours 24_hours, s.convenience_store convenience_store, 
       s.distance_from_Miwang_km distance_from_Miwang_km

FROM GAS_STATION
WHERE self = 1 and 24_hours = 1 and convenience_store =1
