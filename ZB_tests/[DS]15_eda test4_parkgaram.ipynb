    - 참고: **pycountry.countries**
<br></br>
        - ISO 3166-2(전 세계 나라 및 부속 영토의 주요 구성 단위의 명칭에 고유 부호(코드)를 부여하는 국제 표준) 기준 국가별 코드를 얻을 수 있는 Python Library
<br></br>
        - 국가명 표기 방식에 따라 잘못된 값 또는 값 검색이 안되는 경우가 많아 주의가 필요함
<br></br>
            - 예시: 대한민국의 경우 'south korea', 'republic of korea', 'korea', 'korea, republic of' 등으로 표기되는데, 이 중 'korea, republic of' 로만 정확한 국가 코드를 얻을 수 있음
            - 혼선을 줄이기 위하여 문제 1-2와 같이 변경할 국가명을 제공함
<br></br>
        - 사용법(상세 사용법은 2.Datas의 참고사항 내 링크 참조)
<br></br>
            - 국가 검색 방법
<br></br>
                - 일반 검색: pycountry.countries.get(name=country_name) -> 하나의 결과값을 return
                - fuzzy 검색: pycountry.countries.search_fuzzy(country_name) -> 하나 이상의 결과값을 list형태로 return
<br></br>
            - 예시
<br></br>
            <img src="images/04-1-3-001.png" style="margin:auto" width=500/> 
<br></br>
            <img src="images/04-1-3-002.png" style="margin:auto" width=600/> 
<br></br>
    - 조건1: df_target에 'code'컬럼을 추가하여 각 row(행)의 국가명에 맞는 2글자 국가코드(alpha_2)를 입력하세요.
<br></br>
    - 조건2: 일반검색(pycountry.countries.get(name=country_name))을 우선 이용 해보고, 결과값이 안 나올 경우 fuzzy 검색(pycountry.countries.search_fuzzy(country_name))을 활용하여 검색하세요.
<br></br>
    - 조건3: fuzzy 검색을 이용할 경우, 결과값 list의 첫번째 값(index=0)의 국가코드를 입력하세요.
<br></br>
    - 조건4: Index 또는 순서(order)는 변경하지 마세요.
<br></br>
<br></br>
    - hint: Python 예외처리(try-except)를 활용해 보세요.
<br></br>
<br></br>
- 완료 후 결과 dataframe 변수를 check_01_03 함수에 입력하여 채점하세요.

```python
import pycountry
```

```python
# pycountry와 for문 이용하여 검색 안되는 국가 확인

country_id = []

for idx, rows in df_target.iterrows():
    country_name = df_target.country[idx]
    country = pycountry.countries.get(name=country_name)
    country_id.append([idx, country])

country_id
```

```python
# 2글자 코드 빼는 법

country_id[0][1].alpha_2
```

```python
# pycountry 일반 검색 안되는 국가 따로 추출

country_none = []

for i in country_id:
    if i[1] == None:
        country_none.append(i)

country_none
```

```python
# fuzzy search
country_id2 = []
for i in country_none:
    country_name = df_target.country[i[0]]
    country = pycountry.countries.search_fuzzy(country_name)
    country_id2.append([i[0], country])

country_id2

# 모두 다중 리스트 형식으로 되어있어서 첫번째만 가져오면 되는 상황에서 굳이 iran 것을 변경할 필요는 없음
```

```python
# code 열 생성
import numpy as np

df_target['code'] = np.nan
df_target.head()
```

```python
# 만들어진 두 리스트를 이용하여 df_target 데이터 프레임에 기입

# pycountry.countries.get 이용한 정보 기입
for i in country_id:
    try:
        df_target.code[i[0]] = i[1].alpha_2
    except:
        continue

# pycountry.countries.fuzzy_search 이용한 정보 기입
for x in country_id2:
    df_target.code[x[0]] = x[1][0].alpha_2

df_target.head()
```

```python
# groupby
grouped = df.groupby(['region', 'sub_region'])

weighted_avg_func1 = lambda g:np.average(g['internet_use_rate'], weights=g['population'])
weighted_avg_func2 = lambda g:np.average(g['income_per_person'], weights=g['population'])


df_test = df.groupby(['region', 'sub_region']).mean()
df_test['weighted_ave_internet'] = grouped.apply(weighted_avg_func1)
df_test['weighted_ave_income'] = grouped.apply(weighted_avg_func2)


df_result = df_test[['weighted_ave_internet', 'weighted_ave_income']]
df_result
```
